```python
from raphtory import Graph
import networkx as nx
import matplotlib.pyplot as plt

g = Graph()

# IDs como strings
g.add_node(timestamp=1, id="A")
g.add_node(timestamp=1, id="B")
g.add_node(timestamp=1, id="C")

g.add_edge(timestamp=1, src="A", dst="B", properties={"status": 0, "latency": 20})
g.add_edge(timestamp=2, src="B", dst="C", properties={"status": 1, "latency": 100})

snapshot = g.at(2)

G_nx = nx.DiGraph()
for node in snapshot.nodes:
    G_nx.add_node(str(node.name))

for edge in snapshot.edges:
    status = edge.properties["status"] if "status" in edge.properties else 0
    latency = edge.properties["latency"] if "latency" in edge.properties else 0
    G_nx.add_edge(edge.src, edge.dst, status=status, latency=latency)

nx.draw(G_nx, with_labels=True)
```
